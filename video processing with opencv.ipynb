```python
 import cv2
#import facevec
import numpy as np
import smtplib

from keras.preprocessing import image 
from keras.models  import load_model

model = load_model('Forest_Forestfire_cnn.h5') 
video = cv2.VideoCapture(0)
name = ['Forest','Forestfire']
    
while(1):
    success, frame = video.read()
    cv2.imwrite("15.jpg",frame)
    img = image.load_img("15.jpg",target_size = (64,64))
    x  = image.img_to_array(img)
    x = np.expand_dims(x,axis = 0)
    pred = model.predict_classes(x)
    p = pred[0]
    print(p[0])
    cv2.putText(frame, "predicted  class = "+str(name[p[0]]), (100,100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 1)
    cv2.imshow("15",frame)
    if cv2.waitKey(1) & 0xFF == ord('a'): 
        break

video.release()
cv2.destroyAllWindows()
```
